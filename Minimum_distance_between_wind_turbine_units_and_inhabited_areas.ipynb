# **Minimum distance allowed between WT units and inhabited areals**
So, this is the main concern while studying wind turbine noise. As noise control regulations are becoming more stringent and less onshore siting are available while the WT units are getting bigger and bigger, the challange is to attend to the regulations by reducing the noise produced without limiting the WT tip speed ratio (TSR) in a way its full potential is not being explored at all. Let us then take a look in some emblematic cases.

By combining the van der Borg model, which relates wind turbine sound power level (SPW) with the rotor diameter, and an equation presented by Sylvio Bistafa relating the SPW to the sound pressure level (SPL) at a given distance r from the source, it is possible to obtain an equation for the minimum distance between WT units and inhabited areas utilizing the local noise control regulations.

## **1 Import libraries**
Oooooooonce again! No excuses if you forget to do it!

```
import numpy as np
import matplotlib
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
```

## **2 Define the range of rotor diameters**
Let us stick with some old to modern technologies ranging from 40 m to 200 m.

```
D_vec = np.linspace(40, 200, 51)
```

## **3 Compute the distance**

```
r_Brazil = 10**(((22*np.log10(D_vec))-40+54)/20) #Same for Germany, Sweden, Australia, Finland and Alberta (Canada)
r_France = 10**(((22*np.log10(D_vec))-35+54)/20)
r_USA_NC = 10**(((22*np.log10(D_vec))-55+54)/20) #Some American states such as North Carolina, Nevada, Michigan and Georgia
r_USA_CO = 10**(((22*np.log10(D_vec))-50+54)/20) #Some American states such as Colorado, Minnesota and Wyoming
r_CAN_ON = 10**(((22*np.log10(D_vec))-45+54)/20) #Same for Vallonia (Belgium), Indiana (USA) and Wisconsin (USA)
r_Den = 10**(((22*np.log10(D_vec))-44+54)/20)
r_BeFl = 10**(((22*np.log10(D_vec))-43+54)/20)
```

## **4 Plot the charts**

```
plt.figure(figsize=(15,7.5))
plt.plot(D_vec, r_Brazil, label="Brazil, Germany, Sweden, Australia, Finland, Alberta (Canada)")
plt.plot(D_vec, r_Den, label="Denmark")
plt.plot(D_vec, r_France, label="France")
plt.plot(D_vec, r_USA_NC, label="Georgia (USA), Michigan (USA), Nevada (USA), North Carolina (USA)")
plt.plot(D_vec, r_USA_CO, label="Colorado (USA), Minnesota (USA), Wyoming (USA)")
plt.plot(D_vec, r_CAN_ON, label="Ontario (Canada), Vallonia (Belgium), Indiana (USA), Wisconsin (USA)")
plt.plot(D_vec, r_BeFl, label="Flanders (Belgium)")
plt.title("Allowed distance between inhabited areas and Wind Turbine sitings based on several countries acoustics regulation")
plt.xlabel('WT rotor diameter [m]')
plt.ylabel('Minimum distance from sitings [m]')
plt.grid(which='major')
plt.grid(which='minor')
plt.legend(loc='upper left')
plt.show()
```
