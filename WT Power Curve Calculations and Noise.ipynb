# **Power output and wind turbine noise**
Based on the last notebook, where we stated how the noise regulations may limit the expansion of bigger WT units, let us then analyze the corresponding power curves and the allowed distances from inhabited areas.

## **1 Import libraries**
Not one word! Just do it!

```
import numpy as np
import matplotlib
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
```

## **2 Define the range of rotor diameters, some constants and pre-caulculations**
Let us stick with some old to modern technologies ranging from 40 m to 200 m.

```
rho = 1.225
D_vec = np.linspace(40, 200, 51)
A_vec = np.pi*(D_vec**2)/4
v = 10
```

## **3 Compute the power output and the distance**

```
P_vec = A_vec*(v**3)*rho*0.5
r_Brazil = 10**(((22*np.log10(D_vec))-40+54)/20) #Same for Germany, Sweden, Australia, Finland and Alberta (Canada)
r_France = 10**(((22*np.log10(D_vec))-35+54)/20)
r_USA_NC = 10**(((22*np.log10(D_vec))-55+54)/20) #Some American states such as North Carolina, Nevada, Michigan and Georgia
r_USA_CO = 10**(((22*np.log10(D_vec))-50+54)/20) #Some American states such as Colorado, Minnesota and Wyoming
r_CAN_ON = 10**(((22*np.log10(D_vec))-45+54)/20) #Same for Vallonia (Belgium), Indiana (USA) and Wisconsin (USA)
r_Den = 10**(((22*np.log10(D_vec))-44+54)/20)
r_BeFl = 10**(((22*np.log10(D_vec))-43+54)/20)
```

## **4 Plot the power output chart**

```
plt.figure(figsize=(7.5,7.5))
plt.plot(D_vec, P_vec/1000000)
plt.title("Wind Turbine power curve for a wind speed of %d m/s" % v)
plt.xlabel('WT rotor diameter [m]')
plt.ylabel('Power [MW]')
plt.grid(which='major')
plt.grid(which='minor')
#plt.legend(loc='upper left')
plt.show()
```

## **5 Plot the power output x allowed distances chart**

```
plt.figure(figsize=(15,7.5))
plt.plot(P_vec/1000000, r_Brazil, label="Brazil, Germany, Sweden, Australia, Finland, Alberta (Canada)")
plt.plot(P_vec/1000000, r_Den, label="Denmark")
plt.plot(P_vec/1000000, r_France, label="France")
plt.plot(P_vec/1000000, r_USA_NC, label="Georgia (USA), Michigan (USA), Nevada (USA), North Carolina (USA)")
plt.plot(P_vec/1000000, r_USA_CO, label="Colorado (USA), Minnesota (USA), Wyoming (USA)")
plt.plot(P_vec/1000000, r_CAN_ON, label="Ontario (Canada), Vallonia (Belgium), Indiana (USA), Wisconsin (USA)")
plt.plot(P_vec/1000000, r_BeFl, label="Flanders (Belgium)")
plt.title("Allowed distance between inhabited areas and Wind Turbine sitings based on several countries acoustics regulation for a wind speed of %d m/s" % v)
plt.xlabel('WT Power output [MW]')
plt.ylabel('Minimum distance from sitings [m]')
plt.grid(which='major')
plt.grid(which='minor')
plt.legend(loc='upper left')
plt.show()
```
